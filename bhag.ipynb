```python
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(y, order=(0, 1, 1))
result = model.fit()

print(result.summary())
```

```python
forecast_steps = 10

forecast = result.get_forecast(steps=forecast_steps)
forecast_mean = forecast.predicted_mean
conf_int = forecast.conf_int()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

# historical data
plt.plot(y, label="Observed")

# forecast
plt.plot(forecast_mean, label="Forecast")

# confidence interval
plt.fill_between(
    conf_int.index,
    conf_int.iloc[:, 0],
    conf_int.iloc[:, 1],
    alpha=0.3,
    label="95% Confidence Interval"
)

plt.xlabel("Year")
plt.ylabel("ANN")
plt.title("ARIMA(0,1,1) Forecast (10 Years)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

forecast_df = pd.DataFrame({
    "Forecast": forecast_mean,
    "Lower_95": conf_int.iloc[:, 0],
    "Upper_95": conf_int.iloc[:, 1]
})

print(forecast_df)
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

model = ExponentialSmoothing(
    y,
    trend="add",
    seasonal=None
)
fit = model.fit()
forecast = fit.forecast(10)

print(forecast)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

# observed data
plt.plot(y, label="Observed", linewidth=2)

# forecasted values
plt.plot(forecast, label="Holt Forecast", linestyle="--", linewidth=2)

plt.xlabel("Year")
plt.ylabel("ANN")
plt.title("Holt’s Linear Trend Forecast")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd

# y = ANN series
t = np.arange(len(y))  # time index

# fit linear trend
coef = np.polyfit(t, y.values, 1)

# future years
future_steps = 10
future_t = np.arange(len(y), len(y) + future_steps)

# forecast
forecast = coef[0] * future_t + coef[1]

# create index
future_years = pd.date_range(
    start=y.index[-1] + pd.DateOffset(years=1),
    periods=future_steps,
    freq="Y"
)

forecast_series = pd.Series(forecast, index=future_years)
print(forecast_series)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(y, label="Observed")
plt.plot(forecast_series, label="Trend Forecast", linestyle="--")
plt.xlabel("Year")
plt.ylabel("Annual Precipitation (ANN)")
plt.title("Projected Annual Precipitation (Bhagirathi Region)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# time index
t = np.arange(len(y))

# fit degree-2 polynomial
coef = np.polyfit(t, y.values, 2)

# future steps
steps = 10
future_t = np.arange(len(y), len(y) + steps)

# forecast values
forecast = coef[0]*future_t**2 + coef[1]*future_t + coef[2]

# future year index
future_years = pd.date_range(
    start=y.index[-1] + pd.DateOffset(years=1),
    periods=steps,
    freq="Y"
)

forecast_series = pd.Series(forecast, index=future_years)

# ----- PLOT -----
plt.figure(figsize=(10, 5))

plt.plot(y, label="Observed", linewidth=2)
plt.plot(forecast_series, label="Polynomial Forecast (Degree 2)",
         linestyle="--", linewidth=2)

plt.xlabel("Year")
plt.ylabel("Annual Precipitation (ANN)")
plt.title("Polynomial Regression Forecast of Annual Precipitation")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

np.random.seed(42)

mu = y.diff().mean()
sigma = y.diff().std()

steps = 10
n_sim = 50

simulations = []

for _ in range(n_sim):
    vals = [y.iloc[-1]]
    for i in range(steps):
        vals.append(vals[-1] + np.random.normal(mu, sigma))
    simulations.append(vals[1:])

future_years = pd.date_range(
    start=y.index[-1] + pd.DateOffset(years=1),
    periods=steps,
    freq="Y"
)

plt.figure(figsize=(10,5))
plt.plot(y, color="black", label="Observed")

for sim in simulations:
    plt.plot(future_years, sim, color="gray", alpha=0.3)

plt.xlabel("Year")
plt.ylabel("ANN Precipitation")
plt.title("Simulated Future Precipitation Scenarios")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd

# historical differences
diff = y.diff().dropna()
mu = diff.mean()
sigma = diff.std()

np.random.seed(42)

n_years = 10
n_sims = 50

future_years = pd.date_range(
    start=y.index[-1] + pd.DateOffset(years=1),
    periods=n_years,
    freq="Y"
)

# DataFrame to store simulations
sim_df = pd.DataFrame(index=future_years)

for i in range(n_sims):
    values = []
    last = y.iloc[-1]
    for _ in range(n_years):
        last = last + np.random.normal(mu, sigma)
        values.append(last)
    sim_df[f"Scenario_{i+1}"] = values
```

```python
print(sim_df.head())
```

```python
summary_df = pd.DataFrame({
    "Dry_10%": sim_df.quantile(0.10, axis=1),
    "Median_50%": sim_df.quantile(0.50, axis=1),
    "Wet_90%": sim_df.quantile(0.90, axis=1)
})

print(summary_df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(y, label="Observed", color="black")
plt.fill_between(
    summary_df.index,
    summary_df["Dry_10%"],
    summary_df["Wet_90%"],
    alpha=0.3,
    label="80% Prediction Band"
)
plt.plot(summary_df["Median_50%"], label="Median Projection", linestyle="--")

plt.legend()
plt.xlabel("Year")
plt.ylabel("ANN Precipitation")
plt.title("Future Precipitation Scenarios (Probabilistic Forecast)")
plt.show()
```

```python
import numpy as np
import pandas as pd

# observed series
obs = y.copy()

# historical increments
diff = obs.diff().dropna()
mu = diff.mean()
sigma = diff.std()

np.random.seed(42)

n_years = len(obs)
n_sims = 100

sim_hist = pd.DataFrame(index=obs.index)

for i in range(n_sims):
    values = [obs.iloc[0]]  # start from first observed year
    for _ in range(1, n_years):
        values.append(values[-1] + np.random.normal(mu, sigma))
    sim_hist[f"Sim_{i+1}"] = values
```

```python
import numpy as np
import pandas as pd

np.random.seed(42)

obs = y.copy()
mu = obs.mean()

n_years = len(obs)
n_sims = 100

sim_hist = pd.DataFrame(index=obs.index)

for i in range(n_sims):
    sim_hist[f"Sim_{i+1}"] = np.random.choice(obs.values, size=n_years, replace=True)
sim_hist.head()
```

```python
obs_mean = obs.mean()
sim_mean = sim_hist.mean().mean()

print("Observed Mean:", obs_mean)
print("Simulated Mean:", sim_mean)
```

```python
obs_var = obs.var()
sim_var = sim_hist.var().mean()

print("Observed Variance:", obs_var)
print("Simulated Variance:", sim_var)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(obs, label="Observed", color="black", linewidth=2)

for i in range(20):  # show only 20 for clarity
    plt.plot(sim_hist.iloc[:, i], color="gray", alpha=0.3)

plt.title("Historical Validation: Observed vs Simulated Precipitation")
plt.xlabel("Year")
plt.ylabel("ANN Precipitation")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.hist(obs, bins=15, density=True, alpha=0.6, label="Observed")
plt.hist(sim_hist.values.flatten(), bins=30, density=True, alpha=0.4, label="Simulated")
plt.legend()
plt.title("Distribution Comparison: Observed vs Simulated")
plt.show()
```
